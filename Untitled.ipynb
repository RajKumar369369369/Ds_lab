```python
import scipy.stats as stats
import pandas as pd
df=pd.read_csv(r"D:\U23AI113\U23AI113\DS_LAB\DS_Lab2\yellow_tripdata_sample.csv")
tips = df['tip_amount'].dropna()

# Perform one-sample t-test against population mean = 2
t_stat, p_value = stats.ttest_1samp(tips, 2)

print("T-statistic:", t_stat)
print("P-value:", p_value)

alpha = 0.05
if p_value < alpha:
    print("Reject H0: The average tip is significantly different from $2")
else:
    print("Fail to Reject H0: No evidence that the average tip differs from $2")
```

```python
import numpy as np
# --- inputs (edit these to match your column names & group labels) ---
value_col  = "fare_amount"
group_col  = "payment_type"
g1, g2     = "Credit card", "Cash"   # put your two groups here (can be codes like 1 and 2)

# --- prepare data ---
d1 = df.loc[(df[group_col]==g1), value_col].dropna()
d2 = df.loc[(df[group_col]==g2), value_col].dropna()

# basic descriptives
n1, n2 = len(d1), len(d2)
m1, m2 = d1.mean(), d2.mean()
s1, s2 = d1.std(ddof=1), d2.std(ddof=1)

# Welch's two-sample t-test (does not assume equal variances)
t_stat, p_val = stats.ttest_ind(d1, d2, equal_var=False)  # two-tailed by default

# (optional) effect size: Hedges' g
sp2 = ((n1-1)*s1**2 + (n2-1)*s2**2) / (n1+n2-2)           # pooled variance
sp  = np.sqrt(sp2)
cohen_d = (m1 - m2) / sp
# small sample correction to Hedges' g
J = 1 - 3/(4*(n1+n2)-9)
hedges_g = J * cohen_d

print(f"Group {g1}: n={n1}, mean={m1:.3f}, sd={s1:.3f}")
print(f"Group {g2}: n={n2}, mean={m2:.3f}, sd={s2:.3f}")
print(f"Welch t-statistic = {t_stat:.3f}, p-value = {p_val:.5f} (two-tailed)")
print(f"Effect size (Hedges' g) = {hedges_g:.3f}")

alpha = 0.05
if p_val < alpha:
    print(f"Reject H0: mean {value_col} differs between {g1} and {g2}.")
else:
    print(f"Fail to reject H0: no evidence of a difference in mean {value_col}.")
```

```python
import pandas as pd
from scipy.stats import chi2_contingency


# Create a contingency table between Payment_type and RateCodeID
contingency_table = pd.crosstab(df["payment_type"], df["RatecodeID"])
print("Contingency Table:")
print(contingency_table)

# Perform Chi-square test
chi2, p, dof, expected = chi2_contingency(contingency_table)

print("\nChi-square Statistic:", chi2)
print("Degrees of Freedom:", dof)
print("p-value:", p)
print("\nExpected Frequencies:")
print(expected)

# Interpretation
alpha = 0.05
if p < alpha:
    print("\n Reject H0: Payment_type and RateCodeID are NOT independent (they are related).")
else:
    print("\n Fail to reject H0: Payment_type and RateCodeID are independent.")
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import pearsonr, spearmanr

# Drop rows with missing values in relevant columns
df = df[['trip_distance', 'fare_amount', 'tip_amount']].dropna()

# 1. Pearson Correlation
pearson_trip_fare, _ = pearsonr(df['trip_distance'], df['fare_amount'])
pearson_fare_tip, _ = pearsonr(df['fare_amount'], df['tip_amount'])

# 2. Spearman Correlation
spearman_trip_fare, _ = spearmanr(df['trip_distance'], df['fare_amount'])
spearman_fare_tip, _ = spearmanr(df['fare_amount'], df['tip_amount'])

print("Pearson correlation (Trip distance vs Fare amount):", pearson_trip_fare)
print("Pearson correlation (Fare amount vs Tip amount):", pearson_fare_tip)
print("Spearman correlation (Trip distance vs Fare amount):", spearman_trip_fare)
print("Spearman correlation (Fare amount vs Tip amount):", spearman_fare_tip)

# 3. Correlation Matrix
corr_matrix = df.corr(method="pearson")

# 4. Heatmap
plt.figure(figsize=(6,4))
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Matrix Heatmap")
plt.show()
```
