```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image

def svd(A):
    ATA = A.T @ A
    eigvals, V = np.linalg.eigh(ATA)  
    
    sorted_idx = np.argsort(eigvals)[::-1]
    eigvals = eigvals[sorted_idx]
    V = V[:, sorted_idx]
    
    singular_vals = np.sqrt(np.maximum(eigvals, 0))
    
    U = np.zeros((A.shape[0], len(singular_vals)))
    for i, sigma in enumerate(singular_vals):
        if sigma > 1e-10:  
            U[:, i] = (A @ V[:, i]) / sigma
    
    return U, singular_vals, V

def load_image(path):
    img = Image.open(path).convert("L")  # grayscale
    return np.array(img, dtype=float)

def svd_reconstruct(U, S, V, k):
    Uk = U[:, :k]
    Sk = np.diag(S[:k])
    Vk = V[:, :k]
    return Uk @ Sk @ Vk.T

def compression_ratio(A, k):
    m, n = A.shape
    return (m*n) / (k*(m+n+1))

paths = [r"D:\PycharmProjects\pythonProject\U23AI113\DS_LAB\DS_Lab4\image1.jpeg",r"D:\PycharmProjects\pythonProject\U23AI113\DS_LAB\DS_Lab4\image2.png",r"D:\PycharmProjects\pythonProject\U23AI113\DS_LAB\DS_Lab4\image3.ppm"]
k_values = [1, 5, 10, 20, 40, 50]

for path in paths:
    A = load_image(path)
    print(f"\nImage: {path}, shape={A.shape}")
    
    U, S, V = svd(A)
    
    plt.figure(figsize=(12, 8))
    plt.suptitle(f"SVD Approximations for {path}", fontsize=14)
    
    plt.subplot(2, 4, 1)
    plt.imshow(A, cmap="gray")
    plt.title("Original")
    plt.axis("off")
    
    for i, k in enumerate(k_values, start=2):
        A_k = svd_reconstruct(U, S, V, k)
        cr = compression_ratio(A, k)
        
        plt.subplot(2, 4, i)
        plt.imshow(A_k, cmap="gray")
        plt.title(f"k={k}\nCR={cr:.2f}")
        plt.axis("off")
    
    plt.tight_layout()
    plt.show()
```

```python
import numpy as np

def pca_reconstruct(Y: np.ndarray, k: int):

    N, D = Y.shape
    mean_vec = Y.mean(axis=0)
    Y_center = Y - mean_vec

    # --- Dual covariance (N x N)
    C = (Y_center @ Y_center.T) / (N - 1)

    # --- Eigen-decomposition of C
    eigvals, eigvecs = np.linalg.eigh(C)            # ascending
    order = np.argsort(eigvals)[::-1]               # descending
    eigvals, eigvecs = eigvals[order], eigvecs[:, order]

    # --- Feature-space eigenvectors (principal directions)
    components = (Y_center.T @ eigvecs) / np.sqrt((N - 1) * eigvals[np.newaxis, :])
    components /= np.linalg.norm(components, axis=0, keepdims=True)  # shape (D, r)

    # --- Take top-k
    components_k = components[:, :k]

    # --- Scores (projection of data)
    scores_k = Y_center @ components_k

    # --- Reconstruction
    Y_recon = mean_vec + scores_k @ components_k.T

    return Y_recon
```
